```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problem 1
c = 0.25
fx = 20
gx = 1
tf = 2
dx = 0.01
dt = dx/1.5

x = np.arange(0,1+dx,dx)
len_x = len(x)
t = np.arange(0,tf+dt,dt)
len_t = len(t)

u = np.zeros((len_x, len_t))

u[:,0] = fx
u[:,1] = gx*dt + fx
u[0,:] = 0
u[-1,:] = 0


r = c**2*dt/(dx)

for n in range(1,len_t-1):
    for i in range(1,len_x-1):
        u[i,n+1] = 2*u[i,n]-u[i,n-1]+(r**2)*(u[i+1,n]-2*u[i,n]+u[i-1,n])

halfsec_index = int(0.5/dt)
onesec_index = int(1/dt)
onehalfsec_index = int(1.5/dt)

plt.plot(x,u[:,halfsec_index],'r',label='0.5s')
plt.plot(x,u[:,onesec_index],'g',label='1s')
plt.plot(x,u[:,onehalfsec_index],'b',label='1.5s')
plt.legend()
plt.title('solutions with $\Delta$t = 0.0066s')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```

```python
#Problem 5
D = 0.002
L = 1
x = np.arange(0,1, 0.005)
n = np.arange(1,3,1)
def func_output(x,t,n=1):
    output = np.zeros((n,len(x)))
    for n in range(1,n):
        output[n-1, :] = (10/ (n*np.pi))* (1-((-1)**n))*np.sin(n*np.pi*x)*np.exp(-(n*np.pi)**2*D*t)
    output = np.sum(output, axis=0)
    return(output)
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot (x, func_output(x,0.2,n=2),label = '2')
plt.plot(x,func_output(x,0.2,n=5),label = '5')
plt.plot(x,func_output(x, 0.2,n=8),label = '8')
plt.plot (x,func_output(x, 0.2, n=10), label = '10')
plt.plot (x,func_output(x, 0.2,n=50), label = '50')
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('Fourier Heat Equation at t = 0.2s')
plt.legend()
plt.subplot(2,1,2)
plt.plot(x,func_output(x,1.5,n=2),label = '2')
plt.plot(x,func_output(x,1.5,n=5), label = '5')
plt.plot(x,func_output(x,1.5,n=8),label = '8')
plt.plot (x,func_output(x,1.5, n=10), label = '10')
plt.plot (x,func_output(x,1.5,n=50), label = '50')
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('Fourier Heat Equation at t = 1.5s')
plt.legend()
plt.show()
```
