```python
import numpy as np
import scipy.integrate as si
import scipy.optimize as so
import matplotlib.pyplot as plt
```

```python
#Problem 3
k = 0.03
l = 1
dx = 0.05

n = np.arange(1,17,1)
f = lambda x: np.tan(x) + 2*x
lam_val = np.zeros((200,1))
for i in range (0,200):
    root = so.fsolve(f,i)
    lam_val[i,0] = root
lam_val = np.unique((np.rint(lam_val*1000))/1000)
lam_new = lam_val[3:]
lam_new = (n-0.5)*np.pi
lam_new = (((np.rint(lam_new*1000))/1000))

def new(x,t,lam,n=1):
    output = np.zeros((n,len(x)))
    for n in range(1,n):
        c = lam[n-1]
        output[n-1,:] = ((10/c)*(1-np.cos(c*l))/0.5*l-(1/(4*c))*np.sin(2*c*l))*np.sin(c*x)*np.exp(-(c)**2*k*t)
    output = np.sum(output,axis=0)
    return(output)

x = np.arange(0,1,dx)
plt.figure(figsize=(10,10))
plt.plot(x,new(x,0.1,lam_new,n=15), label = "0.1 seconds")
plt.plot(x,new(x,0.5,lam_new,n=15), label = "0.5 seconds")
plt.plot(x,new(x,1,lam_new,n=15), label = "1 seconds")
plt.plot(x,new(x,2,lam_new,n=15), label = "2 seconds")
plt.plot(x,new(x,5,lam_new,n=15), label = "5 seconds")
plt.plot(x,new(x,10,lam_new,n=15), label = "10 seconds")
plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.title("Problem Set Lambda Terms")
plt.legend()
plt.show()
```
